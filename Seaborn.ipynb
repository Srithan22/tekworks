```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')

# Load example datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

print('Datasets loaded: tips (rows={}), iris (rows={})'.format(len(tips), len(iris)))

tips.head()
```

```python
sns.histplot(tips['total_bill'], kde=False)
plt.title('Histogram - Distribution of Total Bill')
plt.show()
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'])
plt.title('KDE Plot - Smooth Distribution of Total Bill')
plt.show()
```

```python
# Displot (Histogram + KDE)
sns.displot(tips['total_bill'], kde=True)
plt.title('Displot - Total Bill Distribution')
plt.show()
```

3.Categorical Plots (Why: compare groups/categories)

- **Barplot** (`sns.barplot`): shows aggregate (mean) for categories.
- **Countplot** (`sns.countplot`): shows counts of categorical values.
- **Boxplot** (`sns.boxplot`): shows median, quartiles, and outliers.
- **Violinplot** (`sns.violinplot`): boxplot + KDE.

Use `hue` to split by another categorical variable.

```python
# Bar Plot - Average Tip by Gender
sns.barplot(x='sex', y='tip', data=tips)
plt.title('Bar Plot - Average Tip by Gender')
plt.show()
```

```python
# Count Plot - Frequency of Days
sns.countplot(x='day', data=tips)
plt.title('Count Plot - Frequency of Days')
plt.show()
```

```python
# Box Plot - Total Bill by Day
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot - Total Bill by Day')
plt.show()
```

```python
# Violin Plot - Distribution + Summary
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot - Total Bill by Day')
plt.show()
```

4. Relationship Plots (Why: see how two numeric variables relate)
Scatterplot (sns.scatterplot): point cloud to inspect correlation.
Lineplot (sns.lineplot): trends over an ordered variable.
Relplot (sns.relplot): figure-level interface for relational plots; supports faceting.

```python
# Scatter Plot - Total Bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('Scatter Plot - Total Bill vs Tip')
plt.show()
```

```python
# Line Plot - Tip vs Party Size (useful for trend-like visuals)
sns.lineplot(x='size', y='tip', data=tips)
plt.title('Line Plot - Tip vs Party Size')
plt.show()
```

```python
# Relplot with Facets
sns.relplot(x='total_bill', y='tip', hue='sex', col='day', data=tips)
plt.suptitle('Relplot - Tip vs Total Bill by Day & Gender', y=1.02)
plt.show()
```

Matrix & Correlation Plots (Why: inspect relationships between many variables)
Heatmap (sns.heatmap): visualize correlation matrix or any 2D numerical table.
Clustermap (sns.clustermap): heatmap + hierarchical clustering to find groups.

```python
# Select only numeric columns
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation
corr = iris_num.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title("Heatmap - Correlation of Iris Features")
plt.show()
```

```python
# Select only numeric columns (exclude 'species')
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation matrix
corr = iris_num.corr()

# Create the clustermap
sns.clustermap(corr, cmap='coolwarm', annot=True)
plt.suptitle('Clustermap - Grouped Correlations (Numeric Only)', y=1.02)
plt.show()
```

```python
import pandas as pd

# Load dataset
path = '/content/drive/My Drive/Customer_support_data.csv'
csd = pd.read_csv(path)

# 1. Inspect dataset
print("Initial Shape:", csd.shape)
print("Missing Values:\n", csd.isnull().sum())
print("Duplicate Rows:", csd.duplicated().sum())

# 2. Remove duplicate rows
csd = csd.drop_duplicates()
print("Shape after removing duplicates:", csd.shape)

# 3. Handle missing values
# Numeric columns -> fill with mean
numeric_cols = csd.select_dtypes(include=['int64','float64']).columns
csd[numeric_cols] = csd[numeric_cols].fillna(csd[numeric_cols].mean())

# Categorical columns -> fill with mode
categorical_cols = csd.select_dtypes(include=['object']).columns
for col in categorical_cols:
    csd[col] = csd[col].fillna(csd[col].mode()[0])

# 4. Convert date columns to datetime (if any)
for col in csd.columns:
    if 'date' in col.lower():
        csd[col] = pd.to_datetime(csd[col], errors='coerce')

# 5. Re-check missing values
print("Missing Values after cleaning:\n", csd.isnull().sum())

# 6. Final cleaned dataset shape
print("Final dataset shape:", csd.shape)

# Optional: save cleaned dataset
csd.to_csv('/content/drive/My Drive/Customer_support_data_cleaned.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from scipy.stats import linregress

# Load cleaned dataset
path = '/content/drive/My Drive/Customer_support_data_cleaned.csv'
csd = pd.read_csv(path)

# Convert datetime columns if any
for col in csd.columns:
    if 'date' in col.lower():
        csd[col] = pd.to_datetime(csd[col], errors='coerce')
```

```python
# 1 Distribution of Item_price
plt.figure(figsize=(8,5))
plt.hist(csd['Item_price'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Item Price")
plt.xlabel("Item Price")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.4)
plt.show()

# 2 Boxplot of CSAT Score
plt.figure(figsize=(6,4))
sns.boxplot(y=csd['CSAT Score'], color='lightgreen')
plt.title("Boxplot of CSAT Score")
plt.ylabel("CSAT Score")
plt.show()

# 3 Histogram + Boxplot of connected_handling_time
fig, ax = plt.subplots(2,1,figsize=(8,6))
ax[0].hist(csd['connected_handling_time'], bins=20, color='orange', edgecolor='black')
ax[0].set_title("Distribution of Connected Handling Time")
ax[1].boxplot(csd['connected_handling_time'], patch_artist=True, boxprops=dict(facecolor='gold'))
ax[1].set_ylabel("Handling Time")
plt.tight_layout()
plt.show()

# 4 Pie Chart of CSAT Score
csat_counts = csd['CSAT Score'].value_counts().sort_index()
plt.figure(figsize=(6,6))
plt.pie(csat_counts, labels=csat_counts.index, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("CSAT Score Distribution")
plt.show()

# 5 Binned Bar Chart – Item_price
bins = [0,500,1000,1500,2000,5000,csd['Item_price'].max()]
labels = ['0–500','501–1000','1001–1500','1501–2000','2001–5000','5000+']
csd['Price_Range'] = pd.cut(csd['Item_price'], bins=bins, labels=labels)
range_counts = csd['Price_Range'].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.bar(range_counts.index, range_counts.values, color='teal')
plt.title("Common Price Ranges for Items")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.grid(True, alpha=0.4)
plt.show()
```

```python
# 6 Tickets per channel_name
plt.figure(figsize=(7,5))
csd['channel_name'].value_counts().plot(kind='bar', color='coral')
plt.title("Tickets per Channel Name")
plt.xlabel("Channel")
plt.ylabel("Count")
plt.show()

# 7 Category ticket count (sorted)
plt.figure(figsize=(7,5))
csd['category'].value_counts().sort_values().plot(kind='barh', color='steelblue')
plt.title("Tickets per Category")
plt.xlabel("Count")
plt.show()

# 8 Pie chart of Agent Shift
shift_counts = csd['Agent Shift'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(shift_counts, labels=shift_counts.index, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Distribution of Tickets by Agent Shift")
plt.show()

# 9 Top 10 Sub-category values
top_sub = csd['Sub-category'].value_counts().head(10)
plt.figure(figsize=(8,5))
top_sub.plot(kind='barh', color='orchid')
plt.title("Top 10 Sub-category Values")
plt.xlabel("Ticket Count")
plt.show()

# 10 Tenure Bucket counts
plt.figure(figsize=(7,5))
sns.countplot(x='Tenure Bucket', data=csd, palette='viridis')
plt.title("Tickets by Tenure Bucket")
plt.xticks(rotation=30)
plt.show()
```

```python
# 11 Item_price vs CSAT Score
plt.figure(figsize=(7,5))
sns.scatterplot(x='Item_price', y='CSAT Score', data=csd, color='purple')
plt.title("Item Price vs CSAT Score")
plt.show()

# 12 connected_handling_time vs CSAT Score + trendline
plt.figure(figsize=(7,5))
sns.regplot(x='connected_handling_time', y='CSAT Score', data=csd, scatter_kws={'alpha':0.6})
plt.title("Connected Handling Time vs CSAT Score")
plt.show()

# 13 Density – Hexbin
plt.figure(figsize=(7,5))
plt.hexbin(csd['Item_price'], csd['connected_handling_time'], gridsize=20, cmap='viridis')
plt.colorbar(label='Density')
plt.xlabel("Item Price")
plt.ylabel("Connected Handling Time")
plt.title("Hexbin Plot: Price vs Handling Time")
plt.show()
```

```python
# 14 Avg CSAT per channel_name
plt.figure(figsize=(8,5))
sns.barplot(x='channel_name', y='CSAT Score', data=csd, estimator='mean', ci=None, palette='cool')
plt.title("Average CSAT Score per Channel")
plt.xticks(rotation=30)
plt.show()

# 15 Item_price per category
plt.figure(figsize=(8,5))
sns.boxplot(x='category', y='Item_price', data=csd, palette='Set2')
plt.title("Distribution of Item Price per Category")
plt.xticks(rotation=30)
plt.show()

# 16 Handling Time by Agent Shift
plt.figure(figsize=(7,5))
sns.boxplot(x='Agent Shift', y='connected_handling_time', data=csd, palette='Pastel1')
plt.title("Connected Handling Time by Agent Shift")
plt.show()

# 17 Mean + variability of Item_price across top 5 Product_category
top5_prod = csd['Product_category'].value_counts().head(5).index
subset = csd[csd['Product_category'].isin(top5_prod)]
means = subset.groupby('Product_category')['Item_price'].mean()
stds = subset.groupby('Product_category')['Item_price'].std()

plt.figure(figsize=(8,5))
plt.bar(means.index, means.values, yerr=stds.values, capsize=5, color='gold')
plt.title("Mean ± SD of Item Price across Top 5 Product Categories")
plt.xticks(rotation=30)
plt.show()

# 18 CSAT vs Tenure Bucket
plt.figure(figsize=(7,5))
sns.barplot(x='Tenure Bucket', y='CSAT Score', data=csd, estimator='mean', ci=None, palette='mako')
plt.title("Average CSAT by Tenure Bucket")
plt.show()
```

```python
# 19 Stacked bar – channel_name × category
cross_tab = pd.crosstab(csd['channel_name'], csd['category'])
cross_tab.plot(kind='bar', stacked=True, figsize=(9,6), colormap='tab20')
plt.title("Tickets by Channel and Category")
plt.xlabel("Channel")
plt.ylabel("Count")
plt.legend(title="Category", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# 20 Agent Shift distribution across category
cross_tab2 = pd.crosstab(csd['category'], csd['Agent Shift'])
cross_tab2.plot(kind='bar', figsize=(9,6), colormap='Accent')
plt.title("Agent Shift Distribution Across Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.legend(title="Agent Shift", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
# 21 Tickets per day
tickets_per_day = csd['order_date_time'].dt.date.value_counts().sort_index()
plt.figure(figsize=(9,5))
plt.plot(tickets_per_day.index, tickets_per_day.values, marker='o')
plt.title("Tickets Created Per Day")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 22 Daily average CSAT (rolling)
daily_csat = csd.groupby(csd['order_date_time'].dt.date)['CSAT Score'].mean()
plt.figure(figsize=(9,5))
plt.plot(daily_csat.index, daily_csat.rolling(7).mean(), color='purple')
plt.title("Daily Average CSAT Score (7-Day Rolling)")
plt.xticks(rotation=45)
plt.show()

# 23 Weekly issues
date_cols = ['order_date_time', 'Issue_reported at', 'issue_responded', 'Survey_response_Date']

for col in date_cols:
    csd[col] = pd.to_datetime(csd[col], errors='coerce')

csd = csd.dropna(subset=['Issue_reported at'])
weekly = csd['Issue_reported at'].dt.isocalendar().week.value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.bar(weekly.index, weekly.values, color='lightcoral')
plt.title("Weekly Number of Reported Issues")
plt.xlabel("Week Number")
plt.ylabel("Number of Issues")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# 24 Monthly ticket pattern
monthly = csd['order_date_time'].dt.to_period('M').value_counts().sort_index()
plt.figure(figsize=(9,5))
plt.plot(monthly.index.astype(str), monthly.values, marker='o', color='seagreen')
plt.title("Monthly Ticket Volume")
plt.xticks(rotation=45)
plt.show()

# 25 Ticket volume by channel_name over time
plt.figure(figsize=(9,5))
for ch in csd['channel_name'].unique():
    subset = csd[csd['channel_name'] == ch]
    daily = subset['order_date_time'].dt.date.value_counts().sort_index()
    plt.plot(daily.index, daily.values, label=ch)
plt.title("Ticket Volume by Channel Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 26 Time lag between Issue_reported and issue_responded
csd['Response_Time'] = (csd['issue_responded'] - csd['Issue_reported at']).dt.total_seconds()/3600
plt.figure(figsize=(8,5))
plt.scatter(csd['Issue_reported at'], csd['Response_Time'], alpha=0.6, color='darkorange')
plt.title("Response Time Trend Over Time")
plt.ylabel("Response Time (hours)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 27 Resolved issues over time
resolved = csd['issue_responded'].dt.date.value_counts().sort_index()
plt.figure(figsize=(9,5))
plt.plot(resolved.index, resolved.values, color='teal', marker='o')
plt.title("Resolved Issues Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 28 Peak hour/day
csd['Hour'] = csd['order_date_time'].dt.hour
plt.figure(figsize=(7,5))
plt.bar(csd['Hour'].value_counts().sort_index().index, csd['Hour'].value_counts().sort_index().values, color='plum')
plt.title("Tickets by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()
```
